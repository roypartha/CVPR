# Activation Function

In neural networks, an activation function is a mathematical function that is applied to the output of each neuron in a neural network layer. The purpose of activation functions is to introduce non-linearity into the output of a neuron. Without an activation function, the output of a neuron would be a simple linear function of its input, which would limit the complexity of the functions that the neural network could learn.

Activation functions are typically applied to the output of a linear transformation, which combines the inputs to the neuron with learnable weights and biases. The resulting output is then passed through the activation function, which maps the output to a non-linear range.

# Step function

Step function, also known as Heaviside step function or unit step function, is a mathematical function that maps any real number to either 0 or 1, depending on whether the input is negative or non-negative, respectively.
                               
The step or threshold activation function is a simple activation function. In the year 1943 McCulloch & Pitts [@mcculloch1943logical] introduced the original perceptron and used a step activation function like the following equation



This equation outputs a value of 1.0 for incoming values of 
 (threshold value) or higher and 0 for all other values. Step functions, also known as threshold functions, only return 1 (true) for values that are above the specified threshold, as seen in Figure bellow.


```python
import numpy as np
import matplotlib.pyplot as plt

def step_function(x):
    if x < 0:
        return 0
    elif x == 0:
        return 0.5
    else:
        return 1

# generate some x values
x = np.arange(-5, 5, 0.1)

# compute the y values for the step function
y = [step_function(xi) for xi in x]

# plot the step function
plt.plot(x, y)
plt.title('Step Function')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.show()

```

This code uses the numpy.heaviside() function to implement the step function, and the matplotlib library to plot the function. The x-values to plot are defined using numpy.linspace() function. The resulting plot shows the step function u(x) for the given range of x-values.

# Sigmoid or Logistic Activation Function

The sigmoid function also known as logistic function is considered as the primary choice as an activation function since itâ€™s output exists between (0,1). As a result, it's especially useful in models that require the probability to be predicted as an output. Because the likelihood/probability, of anything, only occurs between 0 and 1, sigmoid turns out to be the best option.

The sigmoid function is commonly used in binary classification problems, where the goal is to predict one of two possible outcomes. It can also be used as an activation function in neural networks, where it is often used in the output layer to produce a probability distribution over multiple classes.

However, the sigmoid function can suffer from the vanishing gradients problem, where the gradient of the function becomes very small for extreme input values, which can slow down the learning process during backpropagation in neural networks.


```python
import matplotlib.pyplot as plt
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.arange(-10, 10, 0.1)
y = sigmoid(x)

plt.plot(x, y)
plt.xlabel('Input')
plt.ylabel('Output')
plt.title('Sigmoid Activation Function')
plt.grid(True)
plt.show()
```

# Rectified Linear Units (ReLU)

Rectified Linear Units (ReLU) is an activation function commonly used in deep learning neural networks. The ReLU function maps any input value less than zero to zero, and any input value greater than or equal to zero to the input value itself. In other words, ReLU is a piecewise linear function that outputs the input directly if positive, and 0 otherwise.

In other words, the activation is simply thresholded at zero

The ReLU activation function is widely used in deep neural networks due to its computational efficiency and ability to overcome the vanishing gradient problem that can occur with sigmoid and tanh functions


```python
import matplotlib.pyplot as plt
import numpy as np

def relu(x):
    return np.maximum(0,x)

x = np.arange(-10, 10, 0.1)
y = relu(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.title('ReLU Activation Function')
plt.grid(True)
plt.show()
```

# Leaky ReLU

Leaky ReLU is a variant of the Rectified Linear Unit (ReLU) activation function. It is designed to address the "dying ReLU" problem, which occurs when a large number of ReLU neurons become inactive and output zero for any input. This can slow down the learning process in a neural network.

The Leaky ReLU function is similar to ReLU, but it adds a small slope to the left of the origin. This slope ensures that neurons do not become completely inactive and can still contribute to the learning process. The slope can be a small fixed value or a learnable parameter.

```python
import matplotlib.pyplot as plt
import numpy as np

def leaky_relu(x, alpha=0.01):
    return np.maximum(alpha*x, x)

x = np.linspace(-10, 10, 1000)
y = leaky_relu(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Leaky ReLU')
plt.grid(True)
plt.show()
```

# Parametric ReLU P-RELU

The Parametric ReLU (PReLU) activation function is a variant of the ReLU function that introduces a learnable parameter to improve its flexibility. In the standard ReLU function, any input less than zero is mapped to zero, while inputs greater than zero are mapped linearly. PReLU allows the negative part of the input to be slightly negative, resulting in a non-zero gradient and a more expressive model.

```python
import matplotlib.pyplot as plt
import numpy as np

def prelu(x, alpha):
    return np.where(x > 0, x, alpha * x)

x = np.arange(-10, 11, 1)
y = prelu(x, 0.1)

plt.plot(x, y)
plt.grid(True)
plt.title("Parametric ReLU (PReLU) Function")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

In the above code, we define the PReLU function as prelu(x, alpha), where x is the input to the function and alpha is a learnable parameter that determines the slope of the negative part of the function.

We then use numpy.where() to return x where x > 0 and alpha * x where x <= 0.

Finally, we plot the function for alpha = 0.1 over the range -10 to 10. The resulting graph shows a linear increase for positive values of x, and a smaller linear increase for negative values of x with slope determined by alpha.

# Softmax

The softmax function is commonly used in the output layer of neural networks to produce probabilities of the different classes. It takes as input a vector of real numbers and normalizes it into a probability distribution, where each value represents the probability of the corresponding class.

The formula for the softmax function is as follows:

it takes a real-valued number and squashes it into range between 0 and 1. Therefore, it is especially used for models where we have to predict the probability as an output. Since probability of anything exists only between the range of 0 and 1, sigmoid is the good choice.



```python
import numpy as np
import matplotlib.pyplot as plt

def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum()

x = np.array([1, 2, 3, 4, 5])
y = softmax(x)

print('Input vector:', x)
print('Output probabilities:', y)

# Plot the output probabilities
labels = ['Class 1', 'Class 2', 'Class 3', 'Class 4', 'Class 5']
plt.bar(labels, y)
plt.title('Softmax Output Probabilities')
plt.xlabel('Class')
plt.ylabel('Probability')
plt.show()
```
